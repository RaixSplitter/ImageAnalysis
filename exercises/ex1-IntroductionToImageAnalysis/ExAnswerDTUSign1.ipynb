```python
from skimage import color, io, measure, img_as_ubyte
from skimage.measure import profile_line
from skimage.transform import rescale, resize
import matplotlib.pyplot as plt
import numpy as np
import pydicom as dicom
import os
```

```python
in_dir = "data"
im_file = "DTUSign1.jpg"
im_path = os.path.join(in_dir, im_file)
```

```python
def img_inspect(img):
    # File info
    print("Image shape: ", img.shape)
    print("Image data type: ", img.dtype)
    print("Image min value: ", img.min())
    print("Image max value: ", img.max())

    # Plot the image
    io.imshow(img)
    plt.axis('off')
    io.show()
```

```python
def histogram_analysis(im):
    im_byte = img_as_ubyte(im)
    y, x, patches = plt.hist(im_byte.ravel(), bins=256)
    max_freq = np.max(y)
    max_freq_bin = np.argmax(y)
    bin_left = x[max_freq_bin]
    bin_right = x[max_freq_bin+1]

    patches[max_freq_bin].set_facecolor('r')
    print(f"Max frequency: {max_freq}")
    print(f"Max frequency bin: {max_freq_bin}")
    print(f"Bin left: {bin_left}")
    print(f"Bin right: {bin_right}")
    plt.show()
```

```python
im = io.imread(im_path, as_gray=False, pilmode = "RGB")

```

```python
img_inspect(im)
```

```python
im_red_component = im[:, :, 0]
im_green_component = im[:, :, 1]
im_blue_component = im[:, :, 2]
img_inspect(im_red_component)
img_inspect(im_green_component)
img_inspect(im_blue_component)
```
